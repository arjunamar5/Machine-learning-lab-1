```python
#A8
import pandas as pd
import numpy as np

# Load and preprocess the thyroid data
thyroid_df = pd.read_excel("C:/Users/Lenovo/Downloads/Lab Session Data.xlsx", sheet_name='thyroid0387_UCI')
thyroid_df.replace('?', np.nan, inplace=True)
print("For thyroid sheet")

# Identify binary columns
binary_columns_thyroid = [col for col in thyroid_df.columns if set(thyroid_df[col].dropna().unique()) <= {'t', 'f'}]
binary_columns_thyroid += [col for col in thyroid_df.columns if set(thyroid_df[col].dropna().unique()) <= {'M', 'F'}]

# Convert binary columns to numeric (1/0)
thyroid_df[binary_columns_thyroid] = thyroid_df[binary_columns_thyroid].replace({'t': 1, 'f': 0, 'M': 1, 'F': 0})

# Calculate Jaccard Coefficient (JC) and Simple Matching Coefficient (SMC) for thyroid data
vector1_thyroid = thyroid_df.loc[0, binary_columns_thyroid].astype(int)
vector2_thyroid = thyroid_df.loc[1, binary_columns_thyroid].astype(int)

f11_thyroid = np.sum((vector1_thyroid == 1) & (vector2_thyroid == 1))
f00_thyroid = np.sum((vector1_thyroid == 0) & (vector2_thyroid == 0))
f10_thyroid = np.sum((vector1_thyroid == 1) & (vector2_thyroid == 0))
f01_thyroid = np.sum((vector1_thyroid == 0) & (vector2_thyroid == 1))

jaccard_coeff_thyroid = f11_thyroid / (f01_thyroid + f10_thyroid + f11_thyroid)
simple_match_coeff_thyroid = (f11_thyroid + f00_thyroid) / (f00_thyroid + f01_thyroid + f10_thyroid + f11_thyroid)

print("Jaccard Coefficient (JC) for Thyroid Data:", jaccard_coeff_thyroid)
print("Simple Matching Coefficient (SMC) for Thyroid Data:", simple_match_coeff_thyroid)

# Load and preprocess the marketing campaign data
marketing_df = pd.read_excel("C:/Users/Lenovo/Downloads/Lab Session Data.xlsx", sheet_name='marketing_campaign')
marketing_df.replace('?', np.nan, inplace=True)
print("For marketing sheet")

# Identify binary columns
binary_columns_marketing = [col for col in marketing_df.columns if set(marketing_df[col].dropna().unique()) <= {0, 1}]

# Calculate Jaccard Coefficient (JC) and Simple Matching Coefficient (SMC) for marketing data
vector1_marketing = marketing_df.loc[0, binary_columns_marketing].astype(int)
vector2_marketing = marketing_df.loc[1, binary_columns_marketing].astype(int)

f11_marketing = np.sum((vector1_marketing == 1) & (vector2_marketing == 1))
f00_marketing = np.sum((vector1_marketing == 0) & (vector2_marketing == 0))
f10_marketing = np.sum((vector1_marketing == 1) & (vector2_marketing == 0))
f01_marketing = np.sum((vector1_marketing == 0) & (vector2_marketing == 1))

jaccard_coeff_marketing = f11_marketing / (f01_marketing + f10_marketing + f11_marketing)
simple_match_coeff_marketing = (f11_marketing + f00_marketing) / (f00_marketing + f01_marketing + f10_marketing + f11_marketing)

print("Jaccard Coefficient (JC) for Marketing Data:", jaccard_coeff_marketing)
print("Simple Matching Coefficient (SMC) for Marketing Data:", simple_match_coeff_marketing)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
thyroid_data = pd.read_excel(r"C:/Users/Lenovo/Downloads/Lab Session Data.xlsx", sheet_name='thyroid0387_UCI')

# Replace '?' with NaN
thyroid_data.replace('?', np.nan, inplace=True)

# Initialize a dictionary to store label encoders for each categorical column
label_encoders_dict = {}

# Encode categorical variables and handle missing values
for cat_column in thyroid_data.select_dtypes(include=['object']).columns:
    if thyroid_data[cat_column].isnull().sum() > 0:
        thyroid_data[cat_column].fillna(thyroid_data[cat_column].mode()[0], inplace=True)
    encoder = LabelEncoder()
    thyroid_data[cat_column] = encoder.fit_transform(thyroid_data[cat_column])
    label_encoders_dict[cat_column] = encoder

# Handle missing values in numeric columns by filling with mean
for num_column in thyroid_data.select_dtypes(include=[np.number]).columns:
    thyroid_data[num_column].fillna(thyroid_data[num_column].mean(), inplace=True)

# Select two vectors (rows) for comparison
vector_1 = thyroid_data.iloc[0].values.reshape(1, -1)
vector_2 = thyroid_data.iloc[1].values.reshape(1, -1)

# Calculate Cosine Similarity between the two vectors
cosine_similarity_score = cosine_similarity(vector_1, vector_2)[0][0]

print("Cosine Similarity:", cosine_similarity_score)
```

```python
#A10
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
file_path = r"C:/Users/Lenovo/Downloads/Lab Session Data.xlsx"
thyroid_data = pd.read_excel(file_path, sheet_name='thyroid0387_UCI')

# Replace '?' with NaN
thyroid_data.replace('?', np.nan, inplace=True)

# Initialize a dictionary to store label encoders for each categorical column
label_encoders_dict = {}

# Encode categorical variables and handle missing values
for cat_column in thyroid_data.select_dtypes(include=['object']).columns:
    if thyroid_data[cat_column].isnull().sum() > 0:
        thyroid_data[cat_column].fillna(thyroid_data[cat_column].mode()[0], inplace=True)
    encoder = LabelEncoder()
    thyroid_data[cat_column] = encoder.fit_transform(thyroid_data[cat_column])
    label_encoders_dict[cat_column] = encoder

# Handle missing values in numeric columns by filling with mean
for num_column in thyroid_data.select_dtypes(include=[np.number]).columns:
    thyroid_data[num_column].fillna(thyroid_data[num_column].mean(), inplace=True)

# Select the first 20 observation vectors
first_20_vectors = thyroid_data.iloc[:20]

# Initialize matrices to store JC, SMC, and COS values
n = len(first_20_vectors)
jc_matrix = np.zeros((n, n))
smc_matrix = np.zeros((n, n))
cos_matrix = np.zeros((n, n))

# Compute JC, SMC, and COS for each pair of vectors
for i in range(n):
    for j in range(n):
        if i != j:
            # Binary comparison for JC and SMC
            vec1 = first_20_vectors.iloc[i].astype(int)
            vec2 = first_20_vectors.iloc[j].astype(int)
            
            f11 = np.sum((vec1 == 1) & (vec2 == 1))
            f00 = np.sum((vec1 == 0) & (vec2 == 0))
            f10 = np.sum((vec1 == 1) & (vec2 == 0))
            f01 = np.sum((vec1 == 0) & (vec2 == 1))
            
            jc_matrix[i, j] = f11 / (f11 + f10 + f01) if (f11 + f10 + f01) != 0 else 0
            smc_matrix[i, j] = (f11 + f00) / (f11 + f00 + f10 + f01)
            
            # Cosine similarity
            cos_matrix[i, j] = cosine_similarity(vec1.values.reshape(1, -1), vec2.values.reshape(1, -1))[0][0]

# Plotting heatmaps for JC, SMC, and COS matrices
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.heatmap(jc_matrix, annot=True, cmap='viridis')
plt.title('Jaccard Coefficient (JC) Heatmap')

plt.subplot(1, 3, 2)
sns.heatmap(smc_matrix, annot=True, cmap='viridis')
plt.title('Simple Matching Coefficient (SMC) Heatmap')

plt.subplot(1, 3, 3)
sns.heatmap(cos_matrix, annot=True, cmap='viridis')
plt.title('Cosine Similarity (COS) Heatmap')

plt.tight_layout()
plt.show()
```
